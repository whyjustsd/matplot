```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=[1,2,3]
y=[2,4,6]
plt.figure(figsize=(5,3),dpi=100)
#plt.plot(x,y, label='2x',color ='red',linewidth = 2, marker='.',markeredgecolor='blue')
plt.title("our first graph",fontdict={'fontsize':20})
plt.plot(x,y,'b^--', label='2x')
x2=np.arange(0,4,0.5)
plt.plot(x2, x2**2,'r', label='x^2')
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])
plt.legend()
plt.savefig('mygraph.png', dpi=100)
plt.show()
```

```python
labels=['A','B','C']
values = [1,4,2]
bars = plt.bar(labels, values)
bars[1].set_hatch('/')
bars[2].set_hatch('*')
bars[0].set_hatch('o')
plt.figure(figsize=(6,4))
plt.show()
```
