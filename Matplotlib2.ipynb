```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv('gas_prices.csv')
plt.figure(figsize=(8, 5))
plt.title('Gas prices over time', fontdict={'fontweight': 'bold', 'fontsize': 22})
#plt.plot(gas.Year, gas.USA, 'b.-', label='United States')
#plt.plot(gas.Year, gas.Canada, 'g.-', label='Canada')
#plt.plot(gas.Year, gas.Japan, 'r.-', label='Japan')
#plt.plot(gas.Year, gas['South Korea'], 'g.-')
countries_to_look_at = ["Australia", "USA","Canada"]
for country in gas:
  if country in countries_to_look_at:
    plt.plot(gas.Year, gas[country], marker='.')
plt.xticks(gas.Year[::3])
plt.xlabel('years')
plt.ylabel('price')
plt.legend()
plt.show()
```
